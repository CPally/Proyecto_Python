```python
import yfinance as yf
import pandas as pd

# Descargar datos de acciones de Tesla utilizando yfinance
tesla = yf.Ticker("TSLA")

# Extraer el historial de datos bursátiles
tesla_data = tesla.history(period="max")

# Restablecer el índice
tesla_data.reset_index(inplace=True)

# Visualizar las primeras cinco filas
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página de ingresos de Tesla en MacroTrends
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Realizar la solicitud HTTP para obtener el contenido de la página
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Buscar todas las tablas en la página
tables = soup.find_all('table')

# Seleccionar la tabla que contiene los ingresos trimestrales
tesla_revenue_table = tables[1]  # La segunda tabla tiene los ingresos trimestrales

# Convertir la tabla a un DataFrame usando pandas
tesla_revenue = pd.read_html(str(tesla_revenue_table))[0]

# Limpiar y formatear los datos
tesla_revenue.columns = ["Date", "Revenue"]  # Renombrar columnas
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(',', '').str.replace('$', '')  # Limpiar valores
tesla_revenue["Revenue"] = pd.to_numeric(tesla_revenue["Revenue"], errors="coerce")  # Convertir a numérico
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])  # Convertir a formato de fecha
tesla_revenue = tesla_revenue.dropna()  # Eliminar filas con valores nulos

# Mostrar las últimas cinco filas del DataFrame
print(tesla_revenue.tail())
```

```python
import matplotlib.pyplot as plt

# Definir la función para graficar
def make_graph(x, y, title, x_label, y_label):
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=title)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid(True)
    plt.show()

# Graficar los datos de Tesla
make_graph(
    x=tesla_stock_data['Date'],
    y=tesla_stock_data['Close'],
    title='Precio de Acciones de Tesla',
    x_label='Fecha',
    y_label='Precio de Cierre (USD)'
)
```

```python
# Graficar los datos de GameStop usando la misma función make_graph
make_graph(
    x=gme_data['Date'], 
    y=gme_data['Close'], 
    title='Precio de Acciones de GameStop', 
    x_label='Fecha', 
    y_label='Precio de Cierre (USD)'
)
```
